CNN model building

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
```

```python
Train_dir='/content/Train'
Test_dir='/content/Test'

train_data=ImageDataGenerator(rescale=1/255.0,horizontal_flip=True,shear_range=0.4)

test_data=ImageDataGenerator(rescale=1/255.0)
```

```python
os.listdir('/content/Train')
```

```python
training_set=train_data.flow_from_directory(Train_dir,target_size=(224,224),batch_size=40,color_mode='rgb',class_mode='categorical')

test_set=test_data.flow_from_directory(Test_dir,target_size=(224,224),batch_size=40,color_mode='rgb',class_mode='categorical')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Conv2D,MaxPool2D,BatchNormalization,Dropout
from tensorflow.keras import optimizers,regularizers
from tensorflow.keras import layers, Model
```

```python
model=Sequential()

model.add(Conv2D(32,kernel_size=(3,3),activation='relu',input_shape=(224,224,3)))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(64,kernel_size=(3,3),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(128,kernel_size=(3,3),activation='relu'))
model.add(Flatten())

model.add(Dense(32,activation='relu'))
model.add(Dense(4,activation='softmax'))

print(model.summary())

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

history=model.fit(training_set,epochs=10,validation_data=test_set, steps_per_epoch=len(training_set),validation_steps=len(test_set))

```

```python
ev=model.evaluate(test_set)
```

TRANSFER LEARNING - XCEPTION


```python
from tensorflow.keras.applications.xception import Xception
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.xception import preprocess_input
```

```python
xcep=Xception(include_top=False,weights='imagenet',input_shape=(224,224,3))
```

```python
for layer in xcep.layers:
  layer.trainable=False
```

```python
x=Flatten()(xcep.output)
prediction=Dense(4,activation='softmax')(x)

m=Model(inputs=xcep.input,outputs=prediction)

m.summary()

m.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
m.fit(training_set,epochs=5,validation_data=test_set)
```

```python
pred=m.evaluate(test_set)
```
